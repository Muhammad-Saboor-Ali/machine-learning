```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder , LabelEncoder , OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('AI_Student_Life_Pakistan_2026.csv')
```

```python
df.head()
```

```python
x=df.drop('Satisfaction_Level' , axis=1)
y=df['Satisfaction_Level']
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y ,test_size=0.2)
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',OrdinalEncoder(categories=[['School','College','University'] , ['Slight Decline','No Change','Improved']]),['Education_Level','Impact_on_Grades']),
    ('tnf2',OneHotEncoder(sparse_output=False,drop='first'),['Gender','City','AI_Tool_Used','Purpose'])
],remainder='passthrough')
le = LabelEncoder()
```

```python
transformer.fit_transform(X_train)
transformer.transform(X_test)
```

```python
le.fit_transform(y_train)
le.transform(y_test)
```

```python
pd.DataFrame(transformer.fit_transform(X_train) , columns=transformer.get_feature_names_out())
```

```python
pd.DataFrame(le.fit_transform(y_train) , columns=['Satisfaction_Level'])
```
