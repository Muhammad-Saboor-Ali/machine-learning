# Matplotlib

- Used for data visualization

1: Line Plot (Trend ke liye)

    Kab use karein?

- Time series data

- Growth ya decline dekhna ho

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,20,15,25,30]

plt.plot(x, y)
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Trend")
plt.show()
```

2: Bar Plot (Category Comparison)
    Kab use karein?

- Different categories compare karni ho

```python
categories = ["IT","HR","Finance"]
values = [50,30,20]

plt.bar(categories, values)
plt.title("Employees per Department")
plt.show()
```

3: Histogram (Distribution dekhne ke liye)
    Kab use karein?

- Continuous data ka spread dekhna ho

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Data Distribution")
plt.show()
```

4: Scatter Plot (Relationship Check)
    Kab use karein?

- 2 variables ka relation dekhna ho

```python
x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x, y)
plt.title("Relationship")
plt.show()
```

5: Pie Chart (Percentage Share)
    Kab use karein?

- Part-to-whole relationship

```python
sizes = [40,30,20,10]
labels = ["IT","HR","Finance","Marketing"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Department Share")
plt.show()
```

6: Box Plot (Outliers Check)

```python
data = [10,20,15,25,30,100]

plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    "Days":["Mon","Tue","Wed","Thus","Fri"],
    "Sales":[12,15,3,18,19]
}
df = pd.DataFrame(data)
plt.plot(df["Days"],df["Sales"])
plt.xlabel("Days")
plt.ylabel("Sales per Unit")
plt.show()
```

```python
import matplotlib.pyplot as plt

data = {}

n = int(input("Kitne students: "))

for i in range(n):
    name = input("Name: ")
    marks = int(input("Marks: "))
    data[name] = marks

x = list(data.keys())
y = list(data.values())

plt.bar(x, y)
plt.show()
```
